### Q-5) What is Distplot? Using Plotly express, plot a distplot.

Distplot is a function in Plotly Express that creates a histogram with a kernel density estimate (KDE) line on top of it. It combines the information from a histogram and a KDE plot to show the distribution of a variable.

```python
import plotly.express as px
tips = px.data.tips()
fig = px.histogram(tips, x="total_bill", nbins=30, marginal="box", color="sex", opacity=0.7)
fig.update_layout(bargap=0.1)
fig.show()
```

```python
import plotly.figure_factory as ff
import numpy as np
np.random.seed(123)
x = np.random.normal(size=1000)
fig = ff.create_distplot([x], ['Random Data'], bin_size=0.2)
fig.show()
```
