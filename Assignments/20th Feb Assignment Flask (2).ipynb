### Q1. Explain GET and POST methods.

GET: Retrieves data from a server by including parameters in the URL. Used for fetching data and considered idempotent.

POST: Submits data to a server by including it in the body of the request. Used for creating or updating data and not considered idempotent.

### Q2. Why is request used in Flask?

The request in Flask is used to access and handle data submitted by clients in HTTP requests. It allows you to retrieve form data, query parameters, headers, cookies, and more, enabling you to process and respond to client requests effectively.

### Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the client's web browser to a different URL. It is commonly used when you want to direct the user to a different page or route within your Flask application or to an external URL. It helps in handling navigation and providing a smooth user experience by automatically sending the client to the specified URL.

### Q4. What are templates in Flask? Why is the render_template() function used?

templates in Flask are files that define the structure and layout of HTML pages. They separate presentation logic from application logic. 

The render_template() function is used to process these templates and generate HTML responses. It combines the template file with provided data to create dynamic web pages, making it easier to maintain and organize code.
