### Q2) What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
### import numpy as np
### np.random.seed(3)
### x = 3 + np.random.normal(0, 2, 50)
### y = 3 + np.random.normal(0, 2, len(x))

Scatter plot is used to observe relationship between variables and it uses dot to represent them.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.xlabel("X axis data")
plt.ylabel("Y axis data")
plt.title("X vs Y graph")
```

### Q3) Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

subplot() is used for creating multiple plot in one current figure. 
subplot() function adds subplot to a current figure at the specified grid position. 

```python
import numpy as np
#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(3,2,1)
plt.plot(x,y)
#For line 2: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(3,2,2)
plt.plot(x,y)
#For line 3: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(3,2,3)
plt.plot(x,y)
#For line 4: 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(3,2,4)
plt.plot(x,y)
```

### Q4) What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar graph is a chart that plots data with rectangular bars representing the total amount of data for that category.
It is generally used for represent categorical data.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

### Q5) What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is the visual representation of the statistical number of a given data set. The number summary includes:
    minimum,maximum,median,first quartile,third quartile.

Box plots are useful as they provide a visual summary of the data to quickly identify mean values, the dispersion of the data set, and signs of skewness.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
```
